# (B) Comparative Analysis of Datasets

**Online Retail II**
- Supported data mining tasks: Frequent itemsets and association rules (course); Sequential pattern mining using purchase sequences over time per customer (external).
- Data quality issues: ~25% of records lack Customer ID, ~2% of transactions (invoices starting with 'C') have negative quantities, non-product entries (e.g., "Postage", "Manual", "Bank Charges") and inconsistent descriptions.
- Algorithmic feasibility: Itemset mining is feasible on single machine with support thresholds, and pruning. Sequential mining is feasible after sessionization preprocessing.
- Bias considerations: Selection bias is present because the data comes from one UK online retailer in a specific product domain, and the customer base includes many business-to-business (B2B) purchasers, so patterns may not generalize to typical consumer retail.
- Ethical considerations: Customer purchasing behavior is sensitive. Although customer IDs are pseudonymous, there is still a risk of re-identification if results are combined with outside information. To reduce risk, results should be reported only in aggregate, without attempting to reveal individual customers or specific businesses.


**GroupLens MovieLens 32M**
- Supported data mining tasks: Graph mining, clustering/embeddings, and itemset mining (course); Deep recommenders (Neural Collaborative Filtering, LightGCN) (external).
- Data quality issues: High sparsity (99.8%+ of the interaction matrix is empty), insufficient historical data for new users or items, and unstructured text tag.
- Algorithmic feasibility: 32M ratings require sparse matrix optimizations or GPU acceleration for deep models; Full graph analysis may require distributed frameworks to handle the scale.
- Bias considerations: Activity bias from requiring users to have at least 20 ratings, excluding casual users; popularity bias where popular movies receive disproportionately more ratings than niche films.
- Ethical considerations: There is a chance for re-identification risk. Taste profiling can infer sensitive personal attributes from movie preferences. Analysis should focus on aggregate trends and avoid individual user profiling.

**Credit Card Fraud Detection**
- Supported data mining tasks: Anomaly detection and classification under extreme imbalance (course); Deep autoencoder-based anomaly detection (external).
- Data quality issues: High class imbalance (only 0.17% of transactions are fraud), features V1-V28 are transformed into PCA components, (loss of semantic interpretability), and "Time" is relative (seconds from dataset start) rather than absolute.
- Algorithmic feasibility: High feasibility on a single machine as the dataset is small (285k rows) and fits in memory. Evaluation requires appropriate metric selection due to the heavy imbalance.
- Bias considerations: Sampling bias as data covers only a specific 2-day window in Europe; label bias where fraud labels depend on the bank's reporting process, meaning some fraud might be undetected.
- Ethical considerations: Financial data is highly sensitive, though PCA masking provides substantial privacy protection. Fairness is a major concern as false positives in a live system could block legitimate users from accessing their funds, causing real-world harm.

## Selected Dataset: Online Retail II

In this section, we perform Exploratory Data Analysis (EDA) on the Online Retail II dataset. Our goal is to understand the data structure, clean it for downstream tasks, and validate the feasibility of our proposed external technique of Sequential Pattern Mining.

### 1. Data Loading and Cleaning
We load the data from both available years (2009-2011) to capture long-term trends and implement a cleaning pipeline to prepare the dataset for analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_and_clean_retail_data(file_path):
    """
    Loads Online Retail II dataset, merges sheets, and performs cleaning.
    
    Cleaning Decisions:
    1. Drop rows with missing 'Customer ID': We need IDs to track sequences for the external technique.
    2. Remove Cancellations: Invoices starting with 'C' are returns. These distort support counts.
    3. Remove Negative/Zero Quantities: Valid sales must have Quantity > 0.
    4. Remove Non-Positive Prices: Prices <= 0 indicate errors, bad debt adjustments, or free samples.
    """
    print(f"Loading {file_path}...")
    
    # Load both sheets to ensure we have the full timeline
    xl = pd.ExcelFile(file_path)
    df1 = xl.parse('Year 2009-2010')
    df2 = xl.parse('Year 2010-2011')
    df = pd.concat([df1, df2], ignore_index=True)
    
    # Standardize column names (some versions have 'Price', others 'UnitPrice')
    if 'UnitPrice' in df.columns:
        df.rename(columns={'UnitPrice': 'Price'}, inplace=True)
    
    initial_rows = len(df)
    
    # 1. Drop missing Customer IDs
    df_clean = df.dropna(subset=['Customer ID']).copy()
    
    # 2. Filter Cancellations (Invoice starts with 'C')
    # Ensure Invoice is string for string operations
    df_clean['Invoice'] = df_clean['Invoice'].astype(str)
    df_clean = df_clean[~df_clean['Invoice'].str.startswith('C')]
    
    # 3. Filter Negative/Zero Quantities
    df_clean = df_clean[df_clean['Quantity'] > 0]
    
    # 4. Filter Non-Positive Prices
    df_clean = df_clean[df_clean['Price'] > 0]
    
    print(f"Initial Rows: {initial_rows}")
    print(f"Cleaned Rows: {len(df_clean)}")
    print(f"Rows Removed: {initial_rows - len(df_clean)}")
    
    return df_clean

# Update this path to your local file location
FILE_PATH = '../dataset_1/online_retail_II.xlsx' 
df_retail = load_and_clean_retail_data(FILE_PATH)

# Tests
# 1) Verify no missing Customer IDs
assert df_retail['Customer ID'].isna().sum() == 0, "Error: Found NaN in Customer ID after cleaning."

# 2) Verify no Cancellations
# Check if any invoice starts with 'C' (case-insensitive check for robustness)
cancellations = df_retail[df_retail['Invoice'].str.upper().str.startswith('C')]
assert len(cancellations) == 0, f"Error: Found {len(cancellations)} cancellation invoices."

# 3) Verify Positive Quantities
assert (df_retail['Quantity'] > 0).all(), "Error: Found non-positive quantities."

# 4) Verify Positive Prices
assert (df_retail['Price'] > 0).all(), "Error: Found non-positive prices."

# 5) Verify DataFrame is not empty
assert len(df_retail) > 0, "Error: DataFrame is empty after cleaning."

print("Data Cleaning Tests Passed.")
```

I loaded the Online Retail II dataset by merging the two separate annual sheets ("Year 2009-2010" and "Year 2010-2011") into a single DataFrame to capture the full transaction history. I then implemented a cleaning pipeline to prepare the data for downstream tasks. Specifically, I removed rows with missing Customer ID because tracking unique user behavior over time is a prerequisite for our planned sequential analysis. I also filtered out cancellation transactions (invoices starting with "C") and negative quantities, as these represent returns that would otherwise distort support counts for frequent itemsets. Finally, I removed rows with `Price <= 0`, as these typically represent bad debt adjustments or free samples rather than genuine customer purchase intent.

We removed 261,822 rows (about 25% of the original data). The vast majority of these were rows missing `Customer ID.` The remaining 805,549 rows are valid transactions that would enable us to conduct sequential user analysis.

### 2. Basket Analysis
We analyze the distribution of baskets (invoices) to understand typical purchasing behavior. We examine the total quantity of items and the count of distinct stock codes per invoice. We also visualize the distribution of distinct items to detect any skew, which is critical for setting support thresholds—if most baskets are small, high support thresholds might filter out too much data.

```python
import matplotlib.pyplot as plt

def analyze_baskets(df):
    """
    Analyzes basket (Invoice) characteristics.
    
    Large baskets increase the computational cost of finding frequent itemsets. 
    Understanding the distribution helps set appropriate max_len or support thresholds.
    """
    # Group by Invoice to simulate "Baskets"
    basket_sizes = df.groupby('Invoice')['Quantity'].sum() # Total items
    basket_item_counts = df.groupby('Invoice')['StockCode'].nunique()
    
    stats = {
        'n_baskets': len(basket_sizes),
        'avg_items': basket_sizes.mean(),
        'median_items': basket_sizes.median(),
        'max_items': basket_sizes.max(),
        'avg_distinct': basket_item_counts.mean()
    }
    
    print(f"Number of Baskets: {stats['n_baskets']}")
    print(f"Avg Items per Basket: {stats['avg_items']:.2f}")
    print(f"Avg Distinct Items: {stats['avg_distinct']:.2f}")

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.hist(basket_item_counts, bins=50, color='skyblue', edgecolor='black', log=True)
    plt.title('Distribution of Distinct Items per Basket (Log Scale)')
    plt.xlabel('Number of Distinct Items')
    plt.ylabel('Frequency (Log)')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
    
    return stats, basket_sizes

stats, basket_sizes = analyze_baskets(df_retail)

# Tests
# 1) Verify basket count matches unique invoices
assert stats['n_baskets'] == df_retail['Invoice'].nunique(), "Error: Basket count mismatch."

# 2) Verify logical statistics
assert stats['avg_items'] > 0, "Error: Average basket size must be positive."
assert stats['max_items'] >= stats['median_items'], "Error: Max items cannot be less than median."

print("Basket Analysis Tests Passed.")
```

This implies that our association rule mining will largely be driven by these smaller and frequent baskets. The large outlier baskets might introduce noise or combinatorial complexity if not handled properly.

### 4. Sparsity Analysis
Retail datasets can be extremely sparse (i.e., most customers have not bought most items). We calculate the global sparsity score and visualize a random sample of the user–item matrix as a heatmap. If the matrix looks too dense, we might have data leakage or duplicate issues. If the matrix looks too sparse, we may need algorithms like FP-Growth and more careful support thresholds, rather than approaches that assume dense structure.

```python
import seaborn as sns
import matplotlib.pyplot as plt

def analyze_sparsity(df, sample_users=50, sample_items=50):
    """
    Calculates global sparsity and visualizes the local structure.
    
    1. Global Sparsity: Justifies using sparse matrix algorithms (Apriori/FP-Growth).
    2. Visualization: Visually confirms the 'empty' nature of the data and checks 
       for local clustering vs. random noise.
    """
    # 1. Global Calculation
    n_invoices = df['Invoice'].nunique()
    n_items = df['StockCode'].nunique()
    
    matrix_size = n_invoices * n_items
    actual_entries = len(df)
    
    sparsity = 1 - (actual_entries / matrix_size)
    
    print(f"Unique Invoices: {n_invoices}")
    print(f"Unique Items: {n_items}")
    print(f"Global Matrix Sparsity: {sparsity:.6%}")
    
    # 2. Visualization (Sampled)
    # We sample because plotting 37k x 4k is impossible/slow
    sampled_invoices = np.random.choice(df['Invoice'].unique(), sample_users, replace=False)
    sampled_items = np.random.choice(df['StockCode'].unique(), sample_items, replace=False)
    
    subset = df[df['Invoice'].isin(sampled_invoices) & df['StockCode'].isin(sampled_items)]
    
    # Create the matrix (1 if item in invoice, 0 otherwise)
    matrix = subset.pivot_table(index='Invoice', columns='StockCode', values='Quantity', aggfunc='count').fillna(0)
    matrix = matrix.applymap(lambda x: 1 if x > 0 else 0)
    
    # Re-index to ensure we show the full grid even if rows/cols are empty
    matrix = matrix.reindex(index=sampled_invoices, columns=sampled_items, fill_value=0)
    
    print(f"\nPlotting sparsity heatmap for random sample of {sample_users} invoices x {sample_items} items...")
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(matrix, cbar=False, cmap="Greys", linewidths=0.1, linecolor='lightgray')
    plt.title(f'Sparsity Heatmap (Sample: {sample_users} Invoices x {sample_items} Items)\nBlack = Interaction, White = Empty')
    plt.xlabel('Items (StockCode)')
    plt.ylabel('Baskets (Invoice)')
    plt.show()
    
    return sparsity

sparsity_score = analyze_sparsity(df_retail)

# Tests
# 1) Verify sparsity is a valid probability (0 to 1)
assert 0 <= sparsity_score <= 1.0, f"Error: Sparsity {sparsity_score} is out of bounds [0,1]."

# 2) Verify sparsity is high (Expected for retail data)
assert sparsity_score > 0.9, f"Warning: Sparsity {sparsity_score} seems suspiciously low for retail data."

print("Sparsity Analysis Tests Passed.")
```

**Your Answer:**

I calculated the global sparsity of the dataset, which resulted in 99.53%. This means that less than 0.5% of the possible customer-item interactions actually exist. To visualize this, I plotted a heatmap of a random 50x50 sample of the matrix. As expected, the heatmap is almost entirely white (empty), with only a few scattered black points representing purchases.

This means we cannot just load a full dense user–item matrix into memory for algorithms like collaborative filtering or clustering. The 99.5% sparsity score dictates that our future implementation of Frequent Itemset Mining must use algorithms optimized for sparse data structures to be computationally efficient.

### 5. Temporal Gap Analysis

Our proposed external technique is Sequential Pattern Mining, which relies on finding recurring purchase sequences across multiple transactions made by the same customer. To validate if this is feasible, we analyze the time gaps between consecutive visits for returning customers. If most customers only visit once, or if gaps are completely random, this technique would fail. We calculate the distribution of these inter-arrival times to confirm sufficient recurrent behavior.

```python
def analyze_temporal_gaps(df):
    """
    Analyzes time gaps between customer visits.
    
    For Sequential Pattern Mining, we need customers to return multiple times.
    If most customers only buy once, sequential mining is not feasible.
    """
    # Ensure Date format
    df = df.copy()
    df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
    
    # Get unique visit times per customer
    visits = df[['Customer ID', 'InvoiceDate', 'Invoice']].drop_duplicates()
    visits = visits.sort_values(['Customer ID', 'InvoiceDate'])
    
    # Shift to find previous visit date
    visits['prev_date'] = visits.groupby('Customer ID')['InvoiceDate'].shift(1)
    
    # Calculate gap in days
    visits['days_diff'] = (visits['InvoiceDate'] - visits['prev_date']).dt.days
    
    # Drop first visits (NaT)
    gaps = visits['days_diff'].dropna()
    
    print(f"Total return visits: {len(gaps)}")
    print(f"Average days between visits: {gaps.mean():.2f}")
    print(f"Median days between visits: {gaps.median():.2f}")
    
    # Plotting
    plt.figure(figsize=(10, 5))
    plt.hist(gaps, bins=50, color='salmon', edgecolor='black')
    plt.title('Distribution of Days Between Customer Visits')
    plt.xlabel('Days Between Visits')
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
    
    return gaps

gaps = analyze_temporal_gaps(df_retail)

# Tests
# 1) Verify gaps are non-negative
# (We sorted by date, so current - prev should always be >= 0)
assert (gaps >= 0).all(), "Error: Found negative temporal gaps (sorting failure)."

# 2) Verify sufficient data for sequential mining
# We need at least some customers to have returned.
assert len(gaps) > 0, "Error: No return visits found. Sequential mining not feasible."

print("Temporal Analysis Tests Passed.")
```

**Your Answer:**

I grouped the transactions by `Customer ID` and sorted them by `InvoiceDate` to calculate the time difference (in days) between consecutive visits. The analysis shows over 31,000 return visits, and confirms that a significant portion of the customer base engages in repeat behavior. The histogram shows a skewed distribution where most return visits happen within 1-2 months (average about 51 days), though some gaps extend much longer.

The substantial number of return visits and the distribution of time gaps provide strong evidence that Sequential Pattern Mining is feasible. We have enough data to model user journeys and transition probabilities between products over time.

**Tests:**

- I verified that all calculated gaps are non-negative, which confirms that the date sorting logic was applied correctly.
- I asserted that the number of gaps is positive to make sure that there is enough repeated visit data to support Sequential Pattern Mining.
